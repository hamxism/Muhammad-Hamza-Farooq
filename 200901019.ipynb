```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
class LinkedList:
  def __init__(self):
    self.head = None
  def beginning(self, new_data):# 	Insertion at the beginning of the list
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
  def end (self, newElement):# 	Insertion at the end of the list
    newNode = Node(newElement)
    if(self.head == None):
      self.head = newNode
      return
    else:
      temp = self.head
      while(temp.next != None):
        temp = temp.next
      temp.next = newNode
  def arbitrary(self, prev_node, new_data):# 	Insertion in arbitrary location of the list
        new_node = Node(new_data)
        new_node.next = prev_node.next
        prev_node.next = new_node
  def delete_beginning(self, key):
        temp = self.head
        if (temp is not None):
            if (temp.data == key):
                self.head = temp.next
                temp = None
                return
        while(temp is not None):
            if temp.data == key:
                break
            prev = temp
            temp = temp.next
        if(temp == None):
            return
        prev.next = temp.next
        temp = None
  def print_beginning(self):
        temp = self.head
        while(temp):
            print (" %d" %(temp.data)),
            temp = temp.next
  def delete_end(self):# 	Deletion at the end of the list
    if(self.head != None):
      if(self.head.next == None):
        self.head = None
      else:
        temp = self.head
        while(temp.next.next != None):
          temp = temp.next
        lastNode = temp.next
        temp.next = None
        lastNode = None
  
  def print_end(self):
    temp = self.head
    if(temp != None):
      print("The list we have :", end=" ")
      while (temp != None):
        print(temp.data, end=" ")
        temp = temp.next
      print()
    else:
      print("The list we attain is empty:")
  def delete_arbitrary(self, dele):# 	Deletion at the arbitrary location of the list
        if self.head is None or dele is None:
            return
        if self.head == dele:
            self.head = dele.next
  def print_arbitrary(self, node):
        while(node is not None):
          print (node.data,)
          node = node.next                
MyList = LinkedList()
MyList.beginning(1)
MyList.beginning(2)

MyList.end (4)
MyList.end (5)
MyList.arbitrary(MyList.head.next, 3)

MyList.print_end()
#insertion ended
#Deletation started
MyList.delete_beginning(2)
print ("\n Deletion at the beginning of the list:")
MyList.print_beginning()

MyList.delete_end()
print ("\n Deletion at the end of the list:")
MyList.print_end()


print ("\n We have the following List Remaining= ",)
MyList.print_arbitrary(MyList.head)

MyList.delete_arbitrary(MyList.head)
MyList.delete_arbitrary(MyList.head.next)
MyList.delete_arbitrary(MyList.head.next)

print ("\n After the Deletion from the arbitrary location of the list, we have:",)
MyList.print_arbitrary(MyList.head)
 
```
